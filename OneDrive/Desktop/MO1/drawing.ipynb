```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.arange(7)
y = np.array([3, 1, 2, 4, 5, 3, 4])
y1 = np.array([5, 4, 5, 3, 4, 2, 5])
```

```python
plt.figure(figsize = (10,10))
plt.plot(x, y)
plt.plot(x,y1)
plt.title("  ", fontsize = 20)
plt.xlabel("Ox")
plt.ylabel("Oy", rotation = 0)

plt.yticks(ticks=np.arange(0,8))
plt.xticks(np.arange(7), labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], rotation = 30)

plt.grid(axis = 'x', color = 'red', linewidth = 0.1)
plt.grid(axis = 'y', color = 'green', linewidth = 0.1)
plt.legend(title = '', labels = ["", " "])
plt.show()
```
